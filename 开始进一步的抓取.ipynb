```python
import requests
from bs4 import BeautifulSoup
import csv
import time


def Urls():
    url = 'https://www.liepin.com/zhaopin/?sfrom=click-pc_homepage-centre_searchbox-search_new&d_sfrom=search_fp&key' \
          '=python '
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')  # 初始化
    try:
        ur = []
        while url != 'javascript:;':
            page = soup.find(class_="pagerbar")
            aa = page.find_all('a')
            url = aa[-3].get('href')
            q = requests.get('https://www.liepin.com' + url)
            soup = BeautifulSoup(q.text, 'html.parser')  # 初始化
            time.sleep(1)
            titles = soup.find_all(class_="job-info")
            for i in titles:
                a = i.find_all('a')
                href = a[0].get('href')
                # print(href)
                ur.append(href)
            # print(ur)
            # print(len(ur))
            return ur
        else:
            print('退出运行。')

    except:
        print('退出')


for url in Urls():
    print(url)
    try:
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')  # 初始化
        title = soup.find(class_="title-info").get_text().split()
        message = soup.find(class_="job-item main-message job-description").get_text(strip=True)
        title.append(message)    
    except:
        url = 'https://www.liepin.com' + url
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')  # 初始化
        title = soup.find(class_="title-info").get_text().split()
        message = soup.find(class_="job-main job-description main-message").get_text(strip=True)
        title.append(message)
    print(url)
    print(title)
    time.sleep(2)
    with open('liepin.csv', 'a', encoding='utf-8') as f:  # 打开并追加写入
        f_csv = csv.writer(f)
        f_csv.writerow(title)  # 一次写入一行信息
```
