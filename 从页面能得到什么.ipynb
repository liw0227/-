```python
import requests
import csv
import time

url = 'https://jiangmen.lianjia.com/zufang/pg2/#contentList'
r = requests.get(url)
r = r.text
from bs4 import BeautifulSoup

soup = BeautifulSoup(r, 'html.parser')  # 初始化
houses = soup.find_all(class_="content__list--item--aside")
# house = soup.find_all(class_="content__list--item--title twoline")
main = soup.find_all(class_="content__list--item--main")
with open('zufang.csv', 'w', encoding='utf-8') as f:  # 打开并追加写入
    f_csv = csv.writer(f)
    for i in range(0, len(houses)):
        title = houses[i].get('title')
        #     title = house[i].get_text(strip=True)
        print(title)
        href = houses[i].get('href')
        print(href)
        area = ((main[i].get_text().replace('/', '')).split())[3]
        print(area)
        mm = ((main[i].get_text().replace('/', '')).split())[4]
        print(mm)
        direction = ((main[i].get_text().replace('/', '')).split())[5]
        print(direction)
        floot = ((main[i].get_text().replace('/', '')).split())[6]
        print(floot)
        hide = ((main[i].get_text().replace('/', '')).split())[7]
        print(hide)
        price = ((main[i].get_text().replace('/', '')).split())[-1]
        print(price)
        money = ((main[i].get_text().replace('/', '')).split())[-2]
        m = money + price
        data = {
            title, href, area, mm, direction, floot, hide, m
        }
        f_csv.writerow(data)  # 一次写入一行信息
        
```
